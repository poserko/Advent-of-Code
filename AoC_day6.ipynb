```python
with open("AoC.txt", 'r') as fp:
    aoc=[f.strip() for f in fp]
```

```python
data=[]
group=[]
for line in aoc:
  if line != '':
    group.append(line)
  else:
    data.append(group)
    group=[]

len(data)
```

```python
all=[]
for i in data:
  join = len(set("".join(i)))
  print(join)
  all.append(join)
len(all)
```

```python
sum(all)
```

```python
u=[]
dups = []
for i in data:
  print(i)
  if len(i) <= 1:
    uniq = len(set("".join(i)))    
    u.append(uniq)
  else:
    all = "".join(i)
    print(all)
    dupes = len(list(set([x for x in all if all.count(x) > 1])))
    print(dupes)
    dups.append(dupes)
    
```

```python
print(sum(dups))
print(sum(u))

print(sum(u) + sum(dups))
```
